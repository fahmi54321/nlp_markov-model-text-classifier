```python
import numpy as np
import matplotlib.pyplot as plt
import string
from sklearn.model_selection import train_test_split
```

```python
input_files = [
    'edgar_allan_poe.txt',
    'robert_frost.txt',
]
```

```python
# collect data into lists
input_texts = []
labels = []

for label, f in enumerate(input_files):
  print(f"{f} corresponds to label {label}")

  for line in open(f):
    line = line.rstrip().lower()
    if line :
      # remove punctuation
      line = line.translate(str.maketrans('','', string.punctuation))

      input_texts.append(line)
      labels.append(label)
```

```python
train_text, test_text, ytrain, ytest = train_test_split(input_texts, labels)
```

```python
# convert text into integer
idx = 1
word2idx = {'<unk>' : 0}

# populate word2idx
for text in train_text :
  tokens = text.split()
  for token in tokens:
    if token not in word2idx:
      word2idx[token] = idx
      idx += 1
```

```python
# convert data into integer format
train_text_int = []
test_text_int = []

for text in train_text:
  tokens = text.split() # tokenization
  line_as_int = [word2idx[token] for token in tokens] # map each token to it's corresponding text
  train_text_int.append(line_as_int)

for text in test_text:
  tokens = text.split()
  line_as_int = [word2idx.get(token, 0) for token in tokens] # get function and will ensure that always return a default value of zero, this corresponds to the unknown token
  test_text_int.append(line_as_int)
```

```python
# initialize A and pi metrices - for both classes

V = len(word2idx) # represent the vocabulary size

# initialized each of these arrays to all ones (the reason : we are going to be using add one smoothing)
# so these ones are the initial fake counts for each initial word and each transition
A0 = np.ones((V,V))
pi0 = np.ones(V)
A1 = np.ones((V,V))
pi1 = np.ones(V)
```

```python
# compute counts for A and pi
def compute_counts(text_as_int, A, pi):
  for tokens in text_as_int:
    last_idx = None
    for idx in tokens:
      if last_idx is None:
        # it's the first word in a sentence
        pi[idx] += 1
      else:
        # the last word exists, so count a transition
        A[last_idx, idx] += 1

      # update last idx
      last_idx = idx
```

```python
compute_counts([t for t, y in zip(train_text_int, ytrain) if y == 0], A0, pi0)
compute_counts([t for t, y in zip(train_text_int, ytrain) if y == 1], A1, pi1)
```

```python
# normalize A and pi so they are valid probability matrices
# convince yourself that this equivalent to the formulas shown before

A0 /= A0.sum(axis = 1, keepdims = True)
pi0 /= pi0.sum()

A1 /= A1.sum(axis = 1, keepdims = True)
pi1 /= pi1.sum()
```

```python
# log A and pi since we don't need the actual probs
logA0 = np.log(A0)
logpi0 = np.log(pi0)

logA1 = np.log(A1)
logpi1 = np.log(pi1)
```

```python
# compute priors
count0 = sum(y == 0 for y in ytrain)
count1 = sum(y == 1 for y in ytrain)
total = len(ytrain)
p0 = count0 / total
p1 = count1 / total
logp0 = np.log(p0)
logp1 = np.log(p1)
p0, p1
```

```python
# build a classifier
class Classifier:
  def __init__(self, logAs, logpis, logpriors):
    self.logAs = logAs
    self.logpis = logpis
    self.logpriors = logpriors
    self.K = len(logpriors) # number of classes

  def _compute_log_likelihood(self, input_, class_):
    logA = self.logAs[class_]
    logpi = self.logpis[class_]

    last_idx = None
    logprob = 0
    for idx in input_:
      if last_idx is None:
        # it's the first token
        logprob += logpi[idx]
      else:
        logprob += logA[last_idx, idx]

      # update last_idx
      last_idx = idx

    return logprob

  def predict(self, inputs):
    predictions = np.zeros(len(inputs))
    for i, input_ in enumerate(inputs):
      posteriors = [self._compute_log_likelihood(input_, c) + self.logpriors[c] \
             for c in range(self.K)]
      pred = np.argmax(posteriors)
      predictions[i] = pred
    return predictions
```

```python
# each array must be in order since classes are assumed to index these lists
clf = Classifier([logA0, logA1], [logpi0, logpi1], [logp0, logp1])
```

```python
ptrain = clf.predict(train_text_int)
print(f"Train acc: {np.mean(ptrain == ytrain)}")
```

```python
ptest = clf.predict(test_text_int)
print(f"Test acc: {np.mean(ptest == ytest)}")
```
